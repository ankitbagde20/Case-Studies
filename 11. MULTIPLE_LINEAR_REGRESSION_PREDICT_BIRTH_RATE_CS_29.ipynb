### Import Libraries

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import Data

```python
os.chdir('C:\\Users\\ankit\\OneDrive\\Documents\\Desktop\\DATA SCIENCE\\Top Mentor Lecture\\24_Day_31_7_21\\Case Studies with Answer\\CS 29 Ans -Multiple Linear Regression')
df = pd.read_excel('Health_Score.xlsx')
df
```

### Creat X

```python
X = df.iloc[:, 1:].values
X
```

### Creat y

```python
y = df.iloc[:, 0].values
y
```

```python
print('Shape of X : ', X.shape,'\n')
print('Shape of y : ', y.shape)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
print('Shape of X train : ', X_train.shape, '\n')
print('Shape of X test : ', X_test.shape, '\n')
print('Shape of y train : ', y_train.shape, '\n')
print('Shape of y train : ', y_test.shape)
```

### Create Linear Regression Model  

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
print('Linear Regration model is fitted successfully')
```

### Prediction

```python
y_pred = lr.predict(X_test)
print('Predicting X test values : \n', y_pred)
```

```python
print('y test values :\n', y_test)
```

### Prediction - Full DataSet

```python
y_pred = lr.predict(X)
print('Predicting full data set : \n', y_pred)
```

### Concatenate the perdition with Actual 

```python
y_pred = pd.DataFrame(y_pred, columns = ['Prediction'])
final = pd.concat([df, y_pred], axis = 1)
display(final)
```

### Coefficients 

```python
print('Coefficients : ',lr.coef_)
```

### Intercept

```python
print('Intercept : ', lr.intercept_)
```
