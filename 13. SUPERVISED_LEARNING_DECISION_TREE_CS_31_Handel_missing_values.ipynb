### 2.	Replace the missing values by the most frequent value in each column. 
[Hint: Refer to Imputer class in Scikit learn preprocessing module]

### Print the number of missing values and Column Data Types 

```python
dtype_null = pd.concat([horse.isnull().sum(), horse.dtypes], axis = 1)
dtype_null.columns = ['Null_count', 'DataType']
dtype_null
```

### columns with missing values and data types - 'Object'

```python
category_col = dtype_null[(dtype_null.Null_count > 0) & (dtype_null.DataType == 'object')]
category_col
```

### Get Column Names asList

```python
category_variables = list(category_col.index)
category_variables
```

### Missing Value Update for Category Column

```python
from sklearn.impute import SimpleImputer

imp = SimpleImputer(strategy = 'most_frequent')
horse[category_variables] = imp.fit_transform(horse[category_variables])
horse
```

### Check missing value count

```python
datatype_null = pd.concat([horse.isnull().sum(), horse.dtypes], axis = 1)
datatype_null.columns = ['Null_count', 'DataType']
datatype_null
```

## 3.	Data  contains many categorical features, replace them with label encoding.

[Hint: Refer to get_dummies methods in pandas dataframe or Label encoder in scikit-learn]


### Get all the columns list data type – Object 

```python
allcategory_col = dtype_null[dtype_null.DataType == 'object']
allcategory_col = list(allcategory_col.index)
allcategory_col
```

### Implement one hot encoding  

```python
horse1 = horse.copy()
```

```python
for i in allcategory_col:
    horse1 = pd.get_dummies(horse1, columns = [i], drop_first = True )
horse1
```

## 4.	Fit a decision tree classifier and observe the accuracy.

### Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(horse1, target, test_size = 0.2, random_state = 1)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### Decision Tree  

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()

dt.fit(X_train, y_train)
```

### Prediction x_test 


```python
y_pred = dt.predict(X_test)
y_pred
```

### Print Accuracy  

```python
from sklearn import metrics

accuracy = metrics.r2_score(y_pred, y_test)
accuracy
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_pred, y_test)
accuracy
```

```python
dt.score(X_test, y_test)
```

```python
dt.score(X_train, y_train)
```
